```python
lavaan = importr('lavaan')
semtools = importr('semTools')

rprint = ro.r('print')
summary = ro.r('summary')
def print_summary(obj):
    return rprint(summary(obj))

rngkind = "L'Ecuyer-CMRG"

def lv_to_df(lv):
    names = lv.names
    names = [nn.replace('.', '_') for nn in names]

    df = pd.DataFrame(lv, index=names, columns=['val'])
    return df

def permmeaseq_to_df(permmeaseq):
    anova_res = lv_to_df(permmeaseq.slots['ANOVA'])
    anova_res.index = [ii+'_anova' for ii in anova_res.index]
    afi_dif = lv_to_df(permmeaseq.slots['AFI.obs'])
    afi_dif.index = [ii+'_diff' for ii in afi_dif.index]
    afi_ps = lv_to_df(permmeaseq.slots['AFI.pval'])
    afi_ps.index = [ii+'_p' for ii in afi_ps.index]
    return pd.concat([anova_res, afi_dif, afi_ps])
```

```python
null_mdl_strs = {}
mdl_strs = {}
for ss, df in item_lookup.groupby('subscale'):
    cln_ss = ss.replace(" ", "_").replace("/", "_").replace("-", "_")
    null_mdl_str = []
    psis = []
    taus = []
    items = []
    for itemn, item in enumerate(df.sort_values('id_ht').htid.values):
        psis.append(f'{item} ~~ c(psi{itemn+1}, psi{itemn+1})*{item}')
        taus.append(f'{item} ~ c(tau{itemn+1}, tau{itemn+1})*{item}')
        items.append(item)
    null_mdl_str = null_mdl_str + psis + taus
    null_mdl_strs[cln_ss] = null_mdl_str
    mdl_strs[cln_ss] = f'{cln_ss} =~' + ' + '.join(items)
```

```python
rread_csv = ro.r("read.csv")
resp = rread_csv('../data/responses_gender.csv', header=True)
ro.globalenv['resp'] = resp
```

```python
perm_stats = []
config_stats = []
mdls = {}
for cln_ss, mdl in null_mdl_strs.items():
    print("================")
    print("================")
    print(cln_ss.upper())
    print("================")
    print("================")
    
    ro.globalenv[cln_ss+'_null'] = ro.StrVector(mdl)
    ro.globalenv[cln_ss+'_mdl'] = mdl_strs[cln_ss]
    fit_null = ro.r(f'cfa({cln_ss}_null, data=resp, group="gender", estimator="WLSMV")')
    fit_config = ro.r(f'cfa({cln_ss}_mdl, data = resp, group = "gender", estimator = "WLSMV")')
    ro.globalenv['fit_config'] = fit_config
    fit_metric = ro.r(f'cfa({cln_ss}_mdl, data = resp, group = "gender", estimator = "WLSMV", group.equal="loadings")')
    fit_scalar = ro.r(f'cfa({cln_ss}_mdl, data = resp, group = "gender", estimator = "WLSMV", group.equal=c("loadings", "intercepts"))')
    fit_strict = ro.r(f'cfa({cln_ss}_mdl, data = resp, group = "gender", estimator = "WLSMV", group.equal=c("loadings", "intercepts", "residuals"))')
    
    myAFIs = ro.StrVector(["chisq", "chisq.scaled", "cfi.robust", "tli.robust", "rmsea.robust"])
    moreAFIs = ro.StrVector(["gammaHat", "adjGammaHat"])
    ro.r(f'set.seed(3141593, kind="{rngkind}")')
    out_config = semtools.permuteMeasEq(nPermute=100, con=fit_config,
                                        AFIs=myAFIs, moreAFIs=moreAFIs,
                                        parallelType="multicore", ncpus=4)
    print("================")
    print("Configural Invariance")
    print("================")
    
    ro.globalenv['fit_config'] = fit_config
    ro.globalenv['myAFIs'] = myAFIs
    ro.r('print(summary(fit_config, fit.measures=TRUE))')
    
    #_ = print_summary(out_config)
    ro.r(f'set.seed(3141593, kind="{rngkind}")')
    out_metric = semtools.permuteMeasEq(nPermute=100, uncon=fit_config, con=fit_metric, null=fit_null,
                                        param="loadings",
                                        AFIs=myAFIs, moreAFIs=moreAFIs,
                                        parallelType="multicore", ncpus=4)
    # print("================")
    # print("Metric Invariance")
    # print("================")
    #_ = print_summary(out_metric)
    
    ro.r(f'set.seed(3141593, kind="{rngkind}")')
    out_scalar = semtools.permuteMeasEq(nPermute=100, uncon=fit_metric, con=fit_scalar, null=fit_null,
                                        param=ro.StrVector(["loadings","intercepts"]),
                                        AFIs=myAFIs, moreAFIs=moreAFIs,
                                        parallelType="multicore", ncpus=4)
    # print("================")
    # print("Scalar Invariance")
    # print("================")
   # _ = print_summary(out_scalar)
    out_strict = semtools.permuteMeasEq(nPermute=100, uncon=fit_scalar, con=fit_strict, null=fit_null,
                                        param=ro.StrVector(["loadings","intercepts", "residuals"]),
                                        AFIs=myAFIs, moreAFIs=moreAFIs,
                                        parallelType="multicore", ncpus=4)
    # print("================")
    # print("Strict Invariance")
    # print("================")
    #_ = print_summary(out_strict)
    
    config_stat = lv_to_df(ro.r('summary(fit_config, fit.measures=TRUE)$fit'))
    config_stat['scale'] = cln_ss
    permconfig_stats = permmeaseq_to_df(out_config)
    permconfig_stats['level'] = 'config'
    permmetric_stats = permmeaseq_to_df(out_metric)
    permmetric_stats['level'] = 'metric'
    permscalar_stats = permmeaseq_to_df(out_scalar)
    permscalar_stats['level'] = 'scalar'
    permstrict_stats = permmeaseq_to_df(out_strict)
    permstrict_stats['level'] = 'strict'
    perm_stat = pd.concat([permconfig_stats, permmetric_stats, permscalar_stats, permstrict_stats])
    perm_stat['scale'] = cln_ss
    print(perm_stat.loc['chisq_p'])

    mdls[cln_ss] = dict(
        config=out_config,
        metric=out_metric,
        scalar=out_scalar,
        strict=out_strict
    )

    config_stats.append(config_stat)
    perm_stats.append(perm_stat)
```
